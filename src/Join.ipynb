# Read in all data sets

### Some Cleanup

```python
# function to drop the punctuations from the string
def drop_pun(str_var):
    ignore='!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    str_cln=str_var.translate({ord(x): '' for x in ignore})
    return str_cln
```

### Regular Season

```python
regular = pd.read_csv('Data/2000_2017_regular_season.csv')
# cleanup asterisks from team names
regular['Team']=regular.Team.apply(drop_pun)
# create a key column
regular['key']=regular[['year', 'Team']].astype(str).apply(lambda x: ''.join(x), axis=1)
regular.head()
```

### Playoff Stats

```python
playoff = pd.read_csv('Data/2000_2017_playoffs.csv')
playoff['key']=playoff[['year', 'Team']].astype(str).apply(lambda x: ''.join(x), axis=1)
```

### Advanced Stats

```python
advanced = pd.read_csv('Data/2007_2018_advanced.csv')
```

```python
# make the a year column
advanced['year'] = ['20'+ x[-2:] for x in advanced.Season]
```
